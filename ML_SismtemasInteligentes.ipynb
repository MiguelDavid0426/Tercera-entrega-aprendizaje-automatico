```python
#Librerias
import cv2
from google.colab.patches import cv2_imshow
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import shutil
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
#Montar google drive
from google.colab import drive
drive.mount('/content/drive')

% cd /content/drive/MyDrive/Colab Notebooks
#Ac√° debemos cargar los archivos de imagenes y etiquetas en google drive

#Cargar etiquetas
lab=np.load('labels.npy')

#Cargar imagenes
img=np.load('Imagenes_gris_60.npy')
```

```python
x_train, x_test, y_train, y_test = train_test_split(img, lab, test_size=0.2)
print(x_train.shape)
print(x_test.shape)
```

```python
train_X = x_train.reshape(-1, 60,60, 1)
test_X = x_test.reshape(-1, 60,60, 1)
train_X.shape, test_X.shape
```

```python
train_X = train_X.astype('float32')
test_X = test_X.astype('float32')
train_X = train_X / 255.
test_X = test_X / 255.
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.5,1,1.5,2],'epsilon':[0.05,0.1,0.15,0.2],'kernel':['linear','poly','rbf','sigmoid']} 

svr=svm.SVR()
model=GridSearchCV(svr,param_grid,scoring='neg_mean_absolute_error')
```

```python
nsamples, nx, ny, nz = train_X.shape
nsamples_t, nx_t, ny_t, nz_t = test_X.shape
```

```python
d2_train_dataset = train_X.reshape((nsamples,nx*ny*nz))
d2_test_dataset = test_X.reshape((nsamples_t,nx_t*ny_t*nz_t))
```

```python
model.fit(d2_train_dataset,y_train)
```

```python
y_pred=model.predict(d2_test_dataset)

#### Error cuadr√°tico medio
print("Error cuadr√°tico medio: ", mean_squared_error(y_test, y_pred)," (the best value is 0)")

#### Puntaje R¬≤, el coeficiente de determinaci√≥n
print("R cuadrado: ", r2_score(y_test, y_pred)," (the best value is 1)")

#### Error absoluto medio
print("MAE: ", mean_absolute_error(y_test, y_pred)," (the best value is 0)")
```
